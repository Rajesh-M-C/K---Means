```python
q1=data["age"].quantile(0.25)
q3=data["age"].quantile(0.75)
iqr=q3-q1
```

```python
df = data[(data["age"]>(q1-1.5*iqr)) & (data["age"]<(q3+1.5*iqr))]
```

```python
df.query('gradyear==2006 and gender=="M"')
```

```python
from sklearn.preprocessing import StandardScaler
ss = df.copy()
names = df.columns[4:]
features = df[names]
features_fit = StandardScaler().fit(features.values)
features_transform = features_fit.transform(features.values)
ss[names] = features_transform
```

```python
ss.head()
```

```python
replace = {"M":1,"F":2,"not disclosed":3}
```

```python
ss["gender"].replace(replace,inplace=True)
```

```python
ss["gender"].value_counts().sort_index()
```

```python
#Building K means model
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(ss)
```

```python
# Elbow method
```

```python
wcss = []
for i in range(1,20,1):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(ss)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,20,1),wcss,"r--")
plt.xlabel("no. of clusters")
plt.ylabel("WCSS")
plt.title("The Elbow plot");
```

```python
kmeans_bm = KMeans(n_clusters=5, random_state=0)
kmeans_bm.fit(ss)
```

```python
kmeans_bm.labels_
```

```python
df["clusters"]=kmeans_bm.labels_
```

```python
pop = df.pop("clusters")
df.insert(4,"clusters",pop)
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
c = ["red","blue","green","cyan","magenta"]
df.groupby("clusters")["age"].count().plot(kind="bar",color=c)
plt.ylabel("No. of Students")
plt.xlabel("Clusters");
```
